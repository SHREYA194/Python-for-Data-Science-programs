## Impute the missing data with all the methods (mean, median and most_frequent)

```python
import pandas as pd
from sklearn.impute import SimpleImputer

car = pd.read_csv("res/cars.csv", sep=';', header=[0,1])

imp_mean = SimpleImputer(strategy='mean')
imp_mean.fit_transform(car)
imp_mean
```

```python
imp_median = SimpleImputer(strategy='median')
imp_median.fit(car)
x = pd.DataFrame(imp_median.transform(car))
x
```

```python
import pandas as pd
car = pd.read_csv("res/cars.csv", sep=";", header=[0,1])
car.fillna(car.mean())
```

```python
import pandas as pd
car = pd.read_csv("res/cars.csv", sep=";", header=[0,1])
car.fillna(car.median())
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer

car = pd.read_csv("res/cars.csv", sep=';', header=[0,1])
imp_mf = SimpleImputer(strategy='most_frequent')
imp_mf.fit(car)
x = pd.DataFrame(imp_mf.transform(car))
x
```
