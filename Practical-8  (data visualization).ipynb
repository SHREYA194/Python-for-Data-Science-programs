## 8.5  Read the data from Topic_Survey_Assignment.csv file and Plot the bar graph. Also Put the percentage values on top of each bar.

```python
#Import needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#Read data from Topic_Survey_Assignment.csv file
topic_survey_assignment = pd.read_csv("res/Topic_Survey_Assignment.csv")
df = pd.DataFrame(topic_survey_assignment)
df.columns = ['Subject', 'Very interested', 'Somewhat interested' ,'Not interested']

barWidth = 0.25

#Divide plot into subplots
fig, ax = plt.subplots(figsize =(20, 10))

very_interested = df['Very interested'].to_list()
somewhat_interested = df['Somewhat interested'].to_list()
not_interested = df['Not interested'].to_list()

#Arrangement and spacing of bars
br1 = np.arange(df.shape[0]) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 

#Plotting bar plot
rects1 = plt.bar(br1, very_interested, color ='r', width = barWidth, edgecolor ='grey') 
rects2 = plt.bar(br2, somewhat_interested, color ='g', width = barWidth, edgecolor ='grey') 
rects3 = plt.bar(br3, not_interested, color ='b', width = barWidth, edgecolor ='grey')

#Add lables to with axis
plt.xlabel('Subject', fontweight ='bold', fontsize=16) 
plt.ylabel('Interest Rate', fontweight ='bold', fontsize=16) 

#Add ticks to axis
plt.xticks([r + barWidth for r in range(len(very_interested))], df['Subject'], fontsize=12)
plt.yticks([])

#Total of interest rate per DataScience Area
tot =  df.sum(axis=1)
total = tot.values.tolist()

#Provides annotation on each bar in the bar plot
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its percentage."""
    for (rect, i) in zip(rects, total) :
        height = rect.get_height()
        ax.annotate('{0:.2f}%'.format((height / i) * 100),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
    i= i + 1

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.legend(['very_interested', 'somewhat_interested', 'not_interested'], loc=0)
plt.title("Percentage of Respondents' Interest in Data Science Areas",fontsize= 20)
plt.show()
```

## 8.7  Plot the Geographic Data from california_cities.csv file using basemap.

```python
import pandas as pd
cities = pd.read_csv('res/california_cities.csv')

# Extract the data we're interested in
lat = cities['latd'].values
lon = cities['longd'].values
population = cities['population_total'].values
area = cities['area_total_km2'].values
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# 1. Draw the map background
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='h', 
            lat_0=37.5, lon_0=-119,
            width=1E6, height=1.2E6)
m.shadedrelief()
m.drawcoastlines(color='gray')
m.drawcountries(color='gray')
m.drawstates(color='gray')

# 2. scatter city data, with color reflecting population
# and size reflecting area
m.scatter(lon, lat, latlon=True,
          c=np.log10(population), s=area,
          cmap='Reds', alpha=0.5)

# 3. create colorbar and legend
plt.colorbar(label=r'$\log_{10}({\rm population})$')
plt.clim(3, 7)

# make legend with dummy points
for a in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.5, s=a,
                label=str(a) + ' km$^2$')
plt.legend(scatterpoints=1, frameon=False,
           labelspacing=1, loc='lower left');

plt.title('Basemap')
plt.show()
```

## 8.8  Here is a Dataset of various Indian cities and the distances between them in edge_list.txt. Draw the Graph for the data.

```python
import networkx as nx

G = nx.read_weighted_edgelist('res/edge_list.txt', delimiter=" ")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize =(15, 10))  

nx.draw_networkx(G, alpha = 0.9, with_labels = True, edge_color ='0.1', cmap = plt.cm.Reds) 

plt.title('Graph - Indian cities - distances')
plt.show()
```
